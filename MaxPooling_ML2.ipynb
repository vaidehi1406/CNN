```python
import numpy as np
from keras.models import Sequential
from keras.layers import MaxPooling2D
```

```python
image = np.array([[2,2,7,3],
                  [9,4,6,1],
                  [8,5,2,4],
                  [3,1,2,6]])
```

```python
image = image.reshape(1,4,4,1)
#(batch_size, height ,width, channels)
```

```python
model = Sequential([MaxPooling2D(pool_size = 2, strides=(2,1))])
#single layer MaxPool2D
#pool_size= (filter size) , stride is 2x2
```

```python
output=model.predict(image)
#generates the pooled output
```

```python
output = np.squeeze(output)
```
